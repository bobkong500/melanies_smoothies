#Add Rows to Flower Detail table

Edit the "set" cells to update the variables, then click "Run All" in the top left-corner.

```
%%sql -r set_root_depth_code
set rdc = 'M';
```

```
%%sql -r set_flower_name
set fn = 'Tulip';
```

```
%%sql -r check_my_variables
select $fn, $rdc;
```
